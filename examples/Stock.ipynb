```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

Now, let's move on to the next step in our plan: downloading the past 10 years of data for AAPL and GOOG stocks. We will download the data from Yahoo finance using the yfinance library.

```python
from datetime import datetime, timedelta

# Define today's date
today = datetime.today().strftime('%Y-%m-%d')

# Define the date 10 years ago
ten_years_ago = (datetime.today() - timedelta(days=10*365)).strftime('%Y-%m-%d')

# Download historical data for AAPL
AAPL = yf.download('AAPL', start=ten_years_ago, end=today)

# Download historical data for GOOG
GOOG = yf.download('GOOG', start=ten_years_ago, end=today)
```

Both AAPL and GOOG stock data for the past 10 years have been successfully downloaded. Let's now proceed with the next part of our plan - plotting the closing price for both the stocks. We will use matplotlib for plotting the data.

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Plot AAPL
ax.plot(AAPL.index, AAPL['Close'], label='AAPL')

# Plot GOOG
ax.plot(GOOG.index, GOOG['Close'], label='GOOG')

# Add title
plt.title('AAPL vs GOOG - Last 10 years')

# Add labels
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
